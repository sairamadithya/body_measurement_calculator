```python
from math import *
print("WELCOME TO THE BODY CALCULATOR!!!")
print("the following body calculations can be made:-")
print(" 1.body adiposity index  2.body mass index  3.body fat percentage  4.body shape index  5.corpulence index  6.lean body mass  7.relative fat mass 8.waist-to-hip ratio  9.waist-to-height ratio")
print("refer this link for ideas on these measurements")
print("https://en.wikipedia.org/wiki/Human_body_weight")
print("enter the measurement to be calculated as it is mentioned above")
choice= input()
if choice== 'body adiposity index':
    print("enter the circumference of the hip in metres")
    hip= float(input())
    print("enter the height in metres")
    height= float(input())
    BAI= 18-((100*hip)/height*sqrt(height))
    print("your body adiposity index is:",BAI)
    print("enter your gender as M or F")
    gender=input()
    if gender=='M':
        print("healthy range is from 13% to 45%")
    else:
        print("healthy range is from 16% to 60%")
elif choice== 'body mass index':
    print("enter the height in metres")
    height= float(input())
    print("enter the weight in kilograms")
    weight= float(input())
    BMI= weight/height**2
    print("your body mass index is:",BMI)
    print("healthy range is from 18.5 to 24.9")
elif choice== 'body fat percentage':
    print('enter the waist circumference in metres')
    waist= float(input())
    print('enter the hip circumference in metres')
    hip=float(input())
    print('enter the neck circumference in metres')
    neck=float(input())
    BFP= waist+hip-neck
    print("your body fat percentage is:",BFP)
    print("healthy range is from 25% to 31%")
elif choice== 'body shape index':
    print('enter the waist circumference in metres')
    waist= float(input())
    print("enter the height in metres")
    height= float(input())
    print("enter the weight in kilograms")
    weight= float(input())
    BMI= float(weight/height**2)
    ABSI= (waist)/(BMI**1.67)*sqrt(height)
    print("you body shape index",ABSI)
elif choice== 'corpulence index':
    print("enter the height in metres")
    height= float(input())
    print("enter the weight in kilograms")
    weight= float(input())
    CI= weight/height**3
    print("your corpulence index is:",CI)
    print("healthy range is from 11 to 15")
elif choice=='lean body mass':
    print('enter the body fat(use the body fat percentage)')
    fat= float(input())
    print("enter the weight in kilograms")
    weight= float(input())
    LBM= weight-fat
    print("your lean body mass is:",LBM)
    print("enter your gender as M or F")
    gender=input()
    if gender=='M':
        print("healthy range is around 90%")
    else:
        print("healthy range is around 70%")
    
elif choice=='relative fat mass':
    print("enter the height in metres")
    height= float(input())
    print('enter the waist circumference in metres')
    waist= float(input())
    print('enter the gender as M or F')
    gender=input()
    if gender=='M':
        RFMM= 64- 20*(height/waist)
        print("your relative fat mass is:",RFMM)
    else:
        RFMF= 76- 20*(height/waist)
        print("the relative fat mass is:",RFMM)
    print("healthy range is from 18.5 to 24.9 (same as BMI)")    
elif choice=='waist-to-hip ratio':
    print('enter the waist circumference in metres')
    waist= float(input())
    print('enter the hip circumference in metres')
    hip= float(input())
    WHR= waist/hip
    print("your waist-to-hip ratio is:",WHR)
    print("enter your gender as M or F")
    gender=input()
    if gender=='M':
        print("healthy range is from 0.96 to 1.00")
    else:
        print("healthy range is from 0.81 to 0.85")
elif choice=='waist-to-height ratio':
    print('enter the waist circumference in metres')
    waist= float(input())
    print("enter the height in metres")
    height= float(input())
    WHR= waist/height
    print("your waist-to-height ratio is:",WHR)
    print("healthy value is around 0.5")
else:
    print("sorry we are unable to calculate the measurement. Sorry for the inconvenience!!")

print("THANK YOU FOR USING THIS APPLICATION!!")
    
```
